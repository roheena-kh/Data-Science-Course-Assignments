```python
import pandas as pd
import matplotlib as plt
```

```python
# reading the data from file
train_data = pd.read_csv("train.csv")
```

```python
# uploading data to data frame
DF = pd.DataFrame(train_data)
DF.head()
```

```python
#checking and handling missing data#
#starting with finding the number of null values
print(DF.isnull().sum())
```

```python
# Hundling missing data

m_age = DF["Age"].median()
DF["Age"] = DF["Age"].fillna(m_age)
DF = DF.dropna(subset=['Embarked'])
#removing Cabin column since it have too many null values
DF = DF.drop(["Cabin"], axis=1)
```

```python
print(DF.isnull().sum())
```

```python
#Filtering instace
female = DF[(DF['Sex'] == 'female')]
print("\nFemale Passengers:")
female
```

```python
#creating new columns 
DF['Date'] = "April 15, 1912"
DF.head(3)
```

```python
# visualizing data as histogram
DF.plot(kind = "hist", figsize=((10, 10)))
```

```python
Age_Sex = DF.groupby("Age")["Sex"].value_counts().unstack()
```

```python
Age_Sex.plot(kind = "bar", figsize=((25, 25)) )
```
